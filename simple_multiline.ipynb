```python
import altair as alt
import pandas as pd
alt.renderers.enable('notebook')
```

```python
df = pd.DataFrame([[x ** 2, x ** 3] for x in range(10)], columns=["squared", "cubed"])
df.index.name = "x"

data = df.reset_index().melt('x')

alt.Chart(data).mark_line().encode(
    x='x',
    y='value',
    color='variable'
)
```
